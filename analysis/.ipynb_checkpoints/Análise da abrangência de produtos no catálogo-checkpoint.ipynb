# Análise da abrangência de produtos no catálogo

## Dúvida da área de negócios
* **Será que nosso catálogo de produtos é abrangente? Ou tem foco em poucas categorias?**

* **Resposta:** Com base na análise realizada o catálogo de produtos é **sim** abrangente. Há 73 categorias de produtos em que pelo menos 45% possui mais de 100 produtos. Um top 10 de categorias com mais produtos foi definido, e o mesmo foi feito para categorias com menos produtos, assim foi possível analisar quais categorias precisam de maior atenção, para que o número de vendas possa ter crescimento.

```python
df_products['product_category_name'].value_counts().plot(kind='bar')
plt.title('Quantida de produtos em cada Categoria')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```

```python
df_products['product_category_name'].value_counts()[:10].plot(kind='bar',color='green')
plt.title('Top 10 de categorias com mais produtos')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```

```python
df_products['product_category_name'].value_counts().sort_values(ascending=True)[:10].plot(kind='bar',color='red')
plt.title('Top 10 de categorias com menos produtos')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```

```python
df_mg_orders_products = df_order_items.merge(df_products, on='product_id')
```

```python
df_mg_orders_products.head()
```

```python
gp = df_mg_orders_products.groupby(by='product_category_name')['price'].sum().to_frame()
gp.reset_index(inplace=True)
```

```python
mean_price_orders_category_name = [go.Bar(x=gp['product_category_name'], y=gp['price'])]

layout = go.Layout(title='Valor de Vendas por Categorias de Produtos')

fig = go.Figure(data=mean_price_orders_category_name, layout=layout)

py.iplot(fig)
```

```python
mean_price_orders_category_name = [go.Bar(x=gp['product_category_name'].head(10), y=gp['price'].)]

layout = go.Layout(title='Top 10 de Categorias que possuem maior valor de venda')

fig = go.Figure(data=mean_price_orders_category_name, layout=layout)

py.iplot(fig)
```

```python
gp.sort_values('price',ascending=True, inplace=True)

mean_price_orders_category_name = [go.Bar(x=gp['product_category_name'].head(10), y=gp['price'].head(10), marker={'color':'#FF0000'})]

layout = go.Layout(title='Top 10 de Categorias que possuem menor valor de venda')

fig = go.Figure(data=mean_price_orders_category_name, layout=layout)

py.iplot(fig)
```
