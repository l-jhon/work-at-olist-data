```python
df_products['product_category_name'].value_counts().plot(kind='bar')
plt.title('Quantida de produtos em cada Categoria')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```

```python
df_products['product_category_name'].value_counts()[:10].plot(kind='bar',color='green')
plt.title('Top 10 de categorias com mais produtos')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```

```python
df_products['product_category_name'].value_counts().sort_values(ascending=True)[:10].plot(kind='bar',color='red')
plt.title('Top 10 de categorias com menos produtos')
plt.ylabel('Quantidade de Produtos', fontsize=12)
plt.xlabel('Categoria de Produtos', fontsize=12)
plt.show()
```
